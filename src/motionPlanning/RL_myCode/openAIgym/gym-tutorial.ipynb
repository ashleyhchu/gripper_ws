# 1. Environments

```python
import gym
env = gym.make('CartPole-v0')
env.reset()

for _ in range(1000):
    
    env.render()
    env.step(env.action_space.sample()) # take a random action
    
env.close()
```

# 2. Observations

```python
import gym
env = gym.make('CartPole-v0')

for i_episode in range(20): #This is just an implementation of the classic “agent-environment loop”.
    
    observation = env.reset() #The process gets started by calling reset(), which returns an initial observation.
    
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action) #The environment’s step function returns exactly what we need.
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
            
env.close()
```

# 3. Spaces

```python
import gym
env = gym.make('CartPole-v0')
print(env.action_space)
#> Discrete(2)
print(env.observation_space)
#> Box(4,)
print(env.observation_space.high)
#> array([ 2.4       ,         inf,  0.20943951,         inf])
print(env.observation_space.low)
#> array([-2.4       ,        -inf, -0.20943951,        -inf])
```

```python
from gym import spaces
space = spaces.Discrete(8) # Set with 8 elements {0, 1, 2, ..., 7}
x = space.sample()
assert space.contains(x)
assert space.n == 8
```
